```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Classroom/Exercise datasets/Ex1/data17.txt', delimiter = "\t", header=None)
```

```python
df
```

```python
df.columns = ['Time','NetTransactionWeight','NetTransactionValue','Change','BuyWeight','BuyValue','SellWeight','SellValue','RoomLeft','RoomOwned']
```

```python
df
```

```python
df[['Code','Date']] = df['Time'].str.split(',', 1, expand=True)
```

```python
df
```

```python
df = df.drop('Time', axis=1)
```

```python
df = df[['Code', 'Date'] + [col for col in df.columns if col not in ['Code', 'Date']]]
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
columns_to_convert = ['NetTransactionWeight','NetTransactionValue', 'BuyWeight', 'BuyValue', 'SellWeight', 'SellValue', 'RoomLeft']
df[columns_to_convert] = df[columns_to_convert].apply(lambda x: pd.to_numeric(x.str.replace(',', '').str.replace('\xa0', ''), errors='coerce'))
```

```python
df.info()
```

```python
df
```

```python
df['Change'] = df['Change'].str.replace(r'\(|\)', '')
```

```python
df['Change'] = df['Change'].str.split(' ').str[0]
```

```python
df['Change'] = df['Change'].astype(float)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
```

```python
df
```

```python
code_list = df['Code'].unique()
```

```python
for code in code_list:
    mask = df['Code'] == code
    df.loc[mask, 'NetTransactionWeight'] = df.loc[mask, 'NetTransactionWeight'].cumsum()

```

```python
for code in code_list:
    mask = df['Code'] == code
    df.loc[mask, 'NetTransactionValue'] = df.loc[mask, 'NetTransactionValue'].cumsum()
```

```python
df
```

```python
for code in code_list:
    mask = df['Code'] == code
    df.loc[mask, 'BuyWeight'] = df.loc[mask, 'BuyWeight'].cumsum()

```

```python
for code in code_list:
    mask = df['Code'] == code
    df.loc[mask, 'BuyValue'] = df.loc[mask, 'BuyValue'].cumsum()

```

```python
for code in code_list:
    mask = df['Code'] == code
    df.loc[mask, 'SellWeight'] = df.loc[mask, 'SellWeight'].cumsum()

```

```python
for code in code_list:
    mask = df['Code'] == code
    df.loc[mask, 'SellValue'] = df.loc[mask, 'SellValue'].cumsum()

```

```python
df['RoomOwned'] = df['RoomOwned'].str.replace(r'%', '')
```

```python
df['RoomOwned'] = df['RoomOwned'].apply(lambda x: pd.to_numeric(str(x).replace(',', '').replace('\xa0', ''), errors='coerce'))
```

```python
df
```

```python
df
```

```python
for code in code_list:
    mask = df['Code'] == code
    df.loc[mask, 'RoomOwned'] = df.loc[mask, 'RoomOwned'].bfill()
```

```python
for code in code_list:
    mask = df['Code'] == code
    df.loc[mask, 'RoomLeft'] = df.loc[mask, 'RoomLeft'].bfill()
```

```python
df
```

```python
df.to_csv('Ex1.csv', index=False)
```
